```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_claims = pd.read_csv('data/Claims.csv')
df_members = pd.read_csv('data/Members.csv')
df_DIH = pd.read_csv('data/DaysInHospital_Y2.csv')
```

```python
def getCleanClaims():
    df = df_claims.loc[df_claims['Year'] == 'Y1']
    df = df.merge(df_DIH, left_on = 'MemberID', right_on = 'MemberID')
    return df
```
